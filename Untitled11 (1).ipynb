```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("C:/newexcel2/student_feedback.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
#data cleaning
df = df.drop(columns=["Row #"], errors="ignore")
```

```python
#handles missing values 
df.isnull().sum()
```

```python
df.fillna(df.mean(numeric_only=True), inplace=True)
```

```python
def simple_sentiment(score):
    if score >= 4:
        return "Positive"
    elif score == 3:
        return "Neutral"
    else:
        return "Negative"

df["Sentiment"] = df["Course recommendation based on relevance"].apply(simple_sentiment)
```

```python
df.head()
```

```python
def sentiment_score_from_rating(rating):
    if rating >= 4:
        return 1.0     # Positive
    elif rating == 3:
        return 0.0     # Neutral
    else:
        return -1.0    # Negative

df["Sentiment_Score"] = df["Course recommendation based on relevance"].apply(
    sentiment_score_from_rating
)
```

```python
def sentiment_label(score):
    if score > 0:
        return "Positive"
    elif score < 0:
        return "Negative"
    else:
        return "Neutral"

df["Sentiment"] = df["Sentiment_Score"].apply(sentiment_label)
```

```python
import matplotlib.pyplot as plt

plt.figure()
df["Sentiment"].value_counts().plot(kind="bar")
plt.title("Sentiment Distribution (Rating-Based)")
plt.xlabel("Sentiment")
plt.ylabel("Number of Students")
plt.show()
```

```python
#satisfaction analysis
parameters = df.columns[1:9]
```

```python
#Calculate Overall Satisfaction
df["Overall_Satisfaction"] = df[parameters].mean(axis=1)
df["Overall_Satisfaction"].head()
```

```python
#Average Rating per Parameter
avg_scores = df[parameters].mean()

plt.figure()
avg_scores.plot(kind="bar")
plt.title("Average Rating per Feedback Parameter")
plt.xlabel("Parameters")
plt.ylabel("Average Rating")
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()
```

```python
#Overall Satisfaction Distribution
plt.figure()
plt.hist(df["Overall_Satisfaction"], bins=5)
plt.title("Overall Satisfaction Distribution")
plt.xlabel("Satisfaction Score")
plt.ylabel("Number of Students")
plt.show()
```

```python
#Top Strengths
top_strengths = avg_scores.sort_values(ascending=False).head(3)
top_strengths
```

```python
plt.figure()
top_strengths.plot(kind="bar")
plt.title("Top 3 Strength Areas")
plt.ylabel("Average Rating")
plt.show()
```

```python
#Areas Needing Improvement
weak_areas = avg_scores.sort_values().head(3)
weak_areas
```

```python
plt.figure()
weak_areas.plot(kind="bar")
plt.title("Areas Needing Improvement")
plt.ylabel("Average Rating")
plt.show()
```

```python
#Course Recommendation Analysis
plt.figure()
df["Course recommendation based on relevance"].value_counts().plot(kind='bar')
plt.title("Course Recommendation Analysis")
plt.xlabel("Recommendation Score")
plt.ylabel("Number of Students")
plt.show()
```
